## Gap Method to Merge Two Sorted Arrays Without Extra Space

```python
def merge(self, nums1, m: int, nums2, n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        start = m
        for i in range(len(nums2)):
            nums1[start] = nums2[i]
            start+=1
        summ = n+m
        if summ%2==0:
            gap = summ//2
        else:
            gap = summ//2
            gap+=1
        while gap!=0:
            l = 0
            r = l+gap
            while r<len(nums1):
                if nums1[l]<nums1[r]:
                    l+=1
                    r+=1
                else:
                    nums1[l], nums1[r] = nums1[r], nums1[l]
                    l+=1
                    r+=1
            if gap==1:
                gap=0
            elif gap%2==0:
                gap = gap//2
            else:
                gap = gap//2
                gap += 1
```

```python
nums1 = [1, 2, 3, 0, 0, 0]
nums2 = [2, 5, 6]
m = 3
n = 3
merge(None, nums1, m, nums2, n)
print(nums1)
```
