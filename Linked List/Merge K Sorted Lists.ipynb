# MERGE K SORTED LISTS USING PRIORITY QUEUE

```python
class Node:
    def __init__(self, value=None):
        self.val = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
    def insertAtEnd(self, value):
        node = Node(value)
        if self.head is None:
            self.head=node
        else:
            itr = self.head
            while itr.next:
                itr=itr.next
            itr.next=node
    def insertAtHead(self, value):
        node=Node(value)
        if self.head is None:
            self.head=node
        else:
            node.next = self.head
            self.head = node
    def deleteLast(self):
        itr = self.head
        if itr is None or itr.next is None:
            self.head=None
        else:
            while itr.next.next:
                itr=itr.next
            itr.next=None
    def deleteHead(self):
        if self.head is None:
            return
        else:
            self.head = self.head.next
    def print(self):
        itr = self.head
        ll = ''
        while itr:
            ll+=str(itr.val)
            ll+=" -> "
            itr=itr.next
        print(ll[:-3])
    def addLL(self, l):
        itr = self.head
        while itr.next:
            itr=itr.next
        itr.next = l.head
    def length(self):
        l = 0
        itr = self.head
        while itr:
            l+=1
            itr=itr.next
        return l

```

```python
ll1 = LinkedList()
ll1.insertAtEnd(10)
ll1.insertAtEnd(12)
ll1.insertAtEnd(34)
ll1.insertAtEnd(43)

ll2 = LinkedList()
ll2.insertAtEnd(16)
ll2.insertAtEnd(17)
ll2.insertAtEnd(28)
ll2.insertAtEnd(39)

ll3 = LinkedList()
ll3.insertAtEnd(6)
ll3.insertAtEnd(17)
ll3.insertAtEnd(18)
ll3.insertAtEnd(19)

ll4 = LinkedList()
ll4.insertAtEnd(11)
ll4.insertAtEnd(33)
ll4.insertAtEnd(35)
ll4.insertAtEnd(40)

ll5 = LinkedList()
ll5.insertAtEnd(61)
ll5.insertAtEnd(18)
ll5.insertAtEnd(8)
ll5.insertAtEnd(7)
```

```python
class Heap:
    def __init__(self):
        self.heap = []
    def compare(self, nodeA, nodeB):
        if nodeA.val < nodeB.val:
            return True
        return False
    def getParent(self, index):
        if index%2==0:
            return index//2 - 1
        return index//2
    def heapify(self, index):
        lowest = index
        left = index*2+1
        right = left+1
        if left<len(self.heap) and self.compare(self.heap[left], self.heap[lowest]):
            lowest = left
        if right<len(self.heap) and self.compare(self.heap[right], self.heap[lowest]):
            lowest = right
        if lowest!=index:
            self.heap[index], self.heap[lowest] = self.heap[lowest], self.heap[index]
            self.heapify(lowest)
    def push(self, node):
        self.heap.append(node)
        index = len(self.heap) - 1
        parent = self.getParent(index)
        while parent>=0 and self.compare(self.heap[index], self.heap[parent]):
            self.heap[parent], self.heap[index] = self.heap[index], self.heap[parent]
            index = parent
            parent = self.getParent(index)

    def pop(self):
        top = self.heap[0]
        self.heap[0], self.heap[-1] = self.heap[-1], self.heap[0]
        self.heap = self.heap[:-1]
        self.heapify(0)
        return top
```

```python
lls = [ll1.head, ll2.head, ll3.head, ll4.head]
heap = Heap()
for head in lls:
    if head:
        heap.push(head)
ans = LinkedList()
ans.insertAtEnd(-1)
itr = ans.head
while heap.heap:
    top = heap.pop()
    if top.next:
        heap.push(top.next)
    itr.next = top
    itr = itr.next
ans.deleteHead()
ans.print()
```
