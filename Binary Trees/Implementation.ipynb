# BINARY TREE IMPLEMENTATION

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
class BT:
    def __init__(self):
        self.root = None
    def insertVal(self, val):
        node = TreeNode(val)
        if self.root is None:
            self.root = node
            return
        queue = [self.root]
        while queue:
            curr = queue.pop(0)
            if curr.left is None:
                curr.left=node
                return
            else:
                queue.append(curr.left)
            if curr.right is None:
                curr.right = node
                return
            else:
                queue.append(curr.right)
    def preOrder(self, node, pre=[]):
        if node is None:
            return None
        pre.append(node.val)
        self.preOrder(node.left, pre)
        self.preOrder(node.right, pre)
        return pre
    def inOrder(self, node, i=[]):
        if node is None:
            return None
        self.inOrder(node.left, i)
        i.append(node.val)
        self.inOrder(node.right, i)
        return i
    def postOrder(self, node, post=[]):
        if node is None:
            return None
        self.postOrder(node.left, post)
        self.postOrder(node.right, post)
        post.append(node.val)
        return post
    def levelOrder(self, node):
        levelTraversal = []
        queue = [node]
        while queue:
            level = []
            for _ in range(len(queue)):
                node = queue.pop(0)
                if node:
                    level.append(node.val)
                    queue.append(node.left)
                    queue.append(node.right)
            if level:
                levelTraversal.append(level)
        return levelTraversal

       
```

# BINARY SEARCH TREE IMPLEMENTATION

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
class BST:
    def __init__(self):
        self.root = None
    def insertVal(self, val):
        node = TreeNode(val)
        if self.root is None:
            self.root = node
            return
        queue = [self.root]
        while queue:
            curr = queue.pop(0)
            if val<curr.val and curr.left is None:
                curr.left=node
                return
            elif val<curr.val:
                queue.append(curr.left)
            if val>curr.val and curr.right is None:
                curr.right = node
                return
            elif val>curr.val:
                queue.append(curr.right)
    def preOrder(self, node, pre=[]):
        if node is None:
            return None
        pre.append(node.val)
        self.preOrder(node.left, pre)
        self.preOrder(node.right, pre)
        return pre
    def inOrder(self, node, i=[]):
        if node is None:
            return None
        self.inOrder(node.left, i)
        i.append(node.val)
        self.inOrder(node.right, i)
        return i
    def postOrder(self, node, post=[]):
        if node is None:
            return None
        self.postOrder(node.left, post)
        self.postOrder(node.right, post)
        post.append(node.val)
        return post
    def levelOrder(self, node):
        levelTraversal = []
        queue = [node]
        while queue:
            level = []
            for _ in range(len(queue)):
                node = queue.pop(0)
                if node:
                    level.append(node.val)
                    queue.append(node.left)
                    queue.append(node.right)
            if level:
                levelTraversal.append(level)
        return levelTraversal
    def findRight(self, node):
        while node.right:
            node=node.right
        return node
    def deleteRoot(self, node):
        if node.left is None:
            return node.right
        elif node.right is None:
            return node.left
        r = self.findRight(node.left)
        r.right = node.right
        return node.left
    def deleteNode(self, key):
        if self.root is None:
            return 
        if self.root.val==key:
            self.root = self.deleteRoot(self.root)
            return
        itr = self.root
        while itr:
            if itr.val>key:
                if itr.left is None:
                    return
                elif itr.left.val==key:
                    itr.left = self.deleteRoot(itr.left)
                    return
                else:
                    itr=itr.left
            else:
                if itr.right is None:
                    return 
                elif itr.right.val==key:
                    itr.right = self.deleteRoot(itr.right)
                    return
                else:
                    itr=itr.right
```

```python
bst = BST()
bst.insertVal(32)
bst.insertVal(34)
bst.insertVal(30)
bst.insertVal(50)
bst.insertVal(33)
bst.insertVal(12)
bst.insertVal(31)
print(bst.preOrder(bst.root, []))
print(bst.inOrder(bst.root, []))
print(bst.postOrder(bst.root, []))
print(bst.levelOrder(bst.root))
bst.deleteNode(32)
print(bst.preOrder(bst.root, []))
print(bst.inOrder(bst.root, []))
print(bst.postOrder(bst.root, []))
print(bst.levelOrder(bst.root))
```
